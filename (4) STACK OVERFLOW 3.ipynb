```python
country_group['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum()).loc['India']
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
country_uses_python
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
python_df = pd.concat([country_respondents,country_uses_python],axis='columns')
python_df
```

```python
python_df.rename(columns={"Country":"NumofRespondents","LanguageWorkedWith":"NumKnowsPython"},inplace=True)
```

```python
python_df['PctKnowsPython'] = round((python_df['NumKnowsPython']/python_df['NumofRespondents']) * 100,2)
```

```python
python_df = python_df.drop('PctofRespondents',axis=1)
```
